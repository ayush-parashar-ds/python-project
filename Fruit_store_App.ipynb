```python
# file_handler.py

import json


STOCK_FILE = "stock.txt"
TRANSACTION_FILE = "transactions.txt"


def load_stock():
    try:
        with open(STOCK_FILE, "r") as file:
            return json.load(file)
    except:
        return {}


def save_stock(stock):
    with open(STOCK_FILE, "w") as file:
        json.dump(stock, file, indent=4)


def save_transaction(transaction):
    with open(TRANSACTION_FILE, "a") as file:
        file.write(transaction + "\n")
```

```python
# logger.py

from datetime import datetime


def write_log(message):
    with open("store.log", "a") as file:
        file.write(
            f"{datetime.now()} - {message}\n"
        )
```

```python
%%writefile file_handler.py

import json

STOCK_FILE = "stock.txt"
TRANSACTION_FILE = "transactions.txt"

def load_stock():
    try:
        with open(STOCK_FILE, "r") as file:
            return json.load(file)
    except:
        return {}

def save_stock(stock):
    with open(STOCK_FILE, "w") as file:
        json.dump(stock, file, indent=4)

def save_transaction(transaction):
    with open(TRANSACTION_FILE, "a") as file:
        file.write(transaction + "\n")
```

```python
%%writefile logger.py

from datetime import datetime

def write_log(message):
    with open("store.log", "a") as file:
        file.write(f"{datetime.now()} - {message}\n")
```

```python
%%writefile fruit_manager.py

from file_handler import load_stock, save_stock
from logger import write_log

def add_fruit():
    stock = load_stock()

    fruit_name = input("Enter Fruit Name: ").title()
    quantity = int(input("Enter Quantity: "))
    price = float(input("Enter Price: "))

    stock[fruit_name] = {
        "quantity": quantity,
        "price": price
    }

    save_stock(stock)
    write_log(f"Added {fruit_name}")

    print("Fruit Added Successfully")
```

```python
%%writefile customer.py

from file_handler import load_stock

def view_stock():
    stock = load_stock()

    if not stock:
        print("No Stock Available")
        return

    for fruit, details in stock.items():
        print(f"{fruit} | Qty={details['quantity']} | Price={details['price']}")
```

```python
%%writefile main.py

from fruit_manager import add_fruit
from customer import view_stock

while True:

    print("\n===== FRUIT STORE =====")
    print("1. Add Fruit")
    print("2. View Stock")
    print("3. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        add_fruit()

    elif choice == "2":
        view_stock()

    elif choice == "3":
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```
